```python
#1. import the neccessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sqlite3
import warnings
warnings.filterwarnings('ignore')
```

```python
#Establish a connection to sqlite database 
conn=sqlite3.connect('movies.sqlite')
#Create a cursor using cursor object
cur=conn.cursor()
#cursor_object.execute('query')
cur.execute('select * from movies')
#fetch the data - fetchall() 
movies=cur.fetchall()
```

```python
movies=pd.DataFrame(movies,columns=['id','original_title','budget','popularity','release_date','revenue','title','vote_average','vote_count',
                                    'overview','tagline','uid','director_id'])
movies
```

```python
cur.execute('select * from directors')
directors=cur.fetchall()
directors=pd.DataFrame(directors,columns=['name','id','gender','uid','department'])
directors
```

```python
#how many movies are present in the data
cur.execute('select count(original_title) from movies')
cur.fetchall()
```

```python
#find the data for James Cameron, Luc Besson and John Woo directors
cur.execute('select * from directors where name in ("James Cameron","Luc Besson","John Woo")')
cur.fetchall()
```

```python
#find all directors with the name starting with 'Steven'
cur.execute('select name from directors where name like "Steven%"')
cur.fetchall()
```

```python
#if 1 is for females then find how many female directors are there
cur.execute('select count(gender) from directors where gender=1')
cur.fetchall()[0][0]
```

```python
#the name of the 10th female director
cur.execute('select name from directors where gender=1 limit 10')   #starting 10 names
cur.fetchall()[-1]  #last value of tuples
#or
cur.execute('select name from directors where gender=1 limit 10')   #starting 10 names
cur.fetchall()[9]   #9th index value of tuples
#or
cur.execute('select name from directors where gender=1 limit 9,1')  #after the 9th value give me 1 value
cur.fetchall()
```

```python
#what are the 3 most popular movies
cur.execute('select original_title from movies order by popularity desc limit 3') 
cur.fetchall()
```

```python
#what is the most awarded average rated movie since jan 1st 2000
cur.execute('select original_title from movies where release_date>"2000-01-01" order by vote_average desc limit 1') 
cur.fetchall()
```

```python
#which movie were directed by Brenda Chapman?
cur.execute('select original_title from movies join directors on movies.director_id=directors.id where name="Brenda Chapman"') 
cur.fetchall()
#or
cur.execute('select original_title from movies,directors where movies.director_id=directors.id and name="Brenda Chapman"') 
cur.fetchall()
```

```python
#name of the director who has made most no of movies
cur.execute('select name,count() from movies join directors on movies.director_id=directors.id group by directors.id order by count(movies.id) desc limit 1') 
cur.fetchall()
#or
cur.execute('select name,count() from movies,directors where movies.director_id=directors.id group by directors.id order by count(movies.id) desc limit 1') 
cur.fetchall()
```

```python
#name of the most bankable director
cur.execute('select name,revenue from movies join directors on movies.director_id=directors.id group by directors.id order by sum(movies.budget) desc limit 1') 
cur.fetchall()
```

```python
#tell everything about top 10 highest budget making movies
cur.execute('select * from movies order by budget desc limit 10') 
pd.DataFrame(cur.fetchall(),columns=movies.columns)
```

```python
#top 10 most revenue making movies
cur.execute('select original_title from movies order by revenue desc limit 10')
cur.fetchall()
```

```python
#name all the directors with all the movies and revenue
cur.execute('select name as "director name",count(movies.title) as "total_movies",sum(movies.revenue) as "total_revenue" from movies join directors on movies.director_id=directors.id group by directors.id order by sum(movies.revenue) desc limit 5')
cur.fetchall()
```

```python
#top 5 directors who has made most flop movies
cur.execute('select name,count(movies.title) as "total",count(revenue<budget) as "total_flop",count(revenue<budget)/count(movies.title) from movies join directors on movies.director_id=directors.id where revenue<budget group by directors.id order by count(movies.id) desc limit 5')
cur.fetchall()
```

```python
#how many male directors have made movies with revenue more than 3227483234
cur.execute('select count(name),sum(revenue) from directors join movies on movies.director_id=directors.id where gender=0 group by directors.id having sum(revenue>3227483234)')   #starting 10 names
cur.fetchall()  #last value of tuples
```

```python
#give title,release date,budget,revenue,popularity and vote average of the movies made by director name like Steven
cur.execute('select original_title,release_date,budget,revenue,popularity,vote_average from directors join movies on movies.director_id=directors.id where directors.name like "Steven%"')
df=cur.fetchall()
df=pd.DataFrame(df,columns=['original_title','release_date','budget','revenue','popularity','vote_average'])
df
```
