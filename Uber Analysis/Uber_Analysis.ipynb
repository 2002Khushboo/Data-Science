```python
df['minutes']=pd.DatetimeIndex(df['END_DATE']-df['START_DATE']).minutes
df
```

```python
df['months']=pd.DatetimeIndex(df['START_DATE']).month
df
```

```python
dic={1:'jan',2:'feb',3:'mar',4:'apr',5:'may',6:'jun',7:'jul',8:'aug',9:'sep',10:'oct',11:'nov',12:'dec'}
df['months']=df['months'].map(dic)
df
```
